See Image 1 (Stories/Tales)

```python
import IPython.display as display
story_tales = '/content/Je55JmRtmkhgmkWf2G5qBsoK71TdJy51Ly0POYrG.jpg'
display.Image(story_tales)
```

```python
# Open the image
from PIL import Image
img = Image.open(story_tales)

# Use pytesseract to extract text, specifying the language code 'tha'
extracted_text = pytesseract.image_to_string(img, lang='tha')

# Print the result
print(extracted_text)
```
